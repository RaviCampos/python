```python
import pandas as pd
pd.set_option('display.float_format', lambda x: '%.2f' % x)
data = pd.read_csv("./kc_house_data.csv")

data.loc[data["bedrooms"] == 1, "dormitory_type"] = "studio"
data.loc[data["bedrooms"] == 2, "dormitory_type"] = "apartament"
data.loc[data["bedrooms"] > 2, "dormitory_type"] = "house"
 
from matplotlib import pyplot as plt

price_by_bedrooms = data[["price", "bedrooms"]].groupby("bedrooms").mean().reset_index()

plt.figure(figsize=(20, 12))
plt.bar( price_by_bedrooms["bedrooms"], price_by_bedrooms["price"])
plt.ticklabel_format(style='plain') 
```

```python
data["level"] = "NA"
```

```python
# give a price level

for i in range(len(data)):
    if data.loc[i, 'price'] < 321950:
        data.loc[i, 'level'] = 'level_0'
    elif data.loc[i, 'price'] < 450000:
        data.loc[i, 'level'] = 'level_1'
    elif data.loc[i, 'price'] < 645000:
        data.loc[i, 'level'] = 'level_2'
    else:
        data.loc[i, 'level'] = 'level_3'
```

```python
# mean price by level

# group by level

price_by_level = data[["price", "level"]].groupby("level").mean().reset_index()
plt.bar(price_by_level["level"], price_by_level["price"])
plt.ticklabel_format(style='plain', axis="y")
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim( user_agent="geoapiExercises")
response = geolocator.reverse("47.5112,-122.257")
response.raw
```

```python
data["place_id"] = "NA"
data["osm_type"] = "NA"
data["country"] = "NA"
data["country_code"] = "NA"

for i in range(15):
    lat = data.loc[i, "lat"]
    lon = data.loc[i, "long"]
    response = geolocator.reverse(f"{lat},{lon}")
    response = response.raw
    if "place_id" in response:
        data.loc[i, "place_id"] = response["place_id"]
    if "osm_type" in response:
        data.loc[i, "osm_type"] = response["osm_type"]
    if "country" in response["address"]:
        data.loc[i, "country"] = response["address"]["country"]
    if "country_code" in response["address"]:
        data.loc[i, "country_code"] = response["address"]["country_code"]
```
